```python
# slide 53

# degree=1 indicates piecewise-linear
transformed_x = dmatrix("bs(sub.year, knots=[1993], degree=1)", 
                        data={'sub.year': sub.year}, return_type='dataframe')

lm_natcrime_sp1 = smf.OLS(sub.n_natcrime, transformed_x).fit()
lm_natcrime_sp1.summary()
```

The `cut` function in Pandas contains the arguments for bins, labels, right, among others. Labels = None by default so we do not have to specify it here

```python
sub['year_cut'] = pd.cut(sub.loc[:, 'year'], bins = 10, right = False, precision = 0)
sub['age_cut'] = pd.cut(sub.loc[:, 'age'], bins = 10, right = False, precision = 0)
sub['cohort_cut'] = pd.cut(sub.loc[:, 'cohort'], bins = 10, right = False, precision = 0)
```

## An everything model

(slide 81)

```python
sub.dtypes
```

Note: The binned variables are currently of the category dtype. However, this raises a `TypeError: data type not understood` in numpy. This is a known issue documented here: https://github.com/pandas-dev/pandas/issues/9581. We can get around this error by converting the binned variables into strings first, then allow them to be fitted as categorical variables using the `C()` notation. 

```python
sub[['age_cut', 'year_cut', 'cohort_cut']] = sub[['age_cut', 'year_cut', 'cohort_cut']].astype(str)
```

```python
lm_natcrime = smf.ols('n_natcrime ~ C(age_cut) + C(year_cut) + C(cohort_cut)', data = sub).fit()
lm_natcrime.summary()
```

```python
# slide 83
sub['year_cut'].unique()
```

## Reduce the whole model just to period effects

```python
# Create a dummy variable `late` for years after 1996 as 1 and otherwise 0
sub['late'] = (sub.year > 1996).astype(int)

lm_natcrime_apc = smf.ols('n_natcrime ~ age + late + C(cohort_cut)', data = sub).fit()
lm_natcrime_apc.summary()
```

Due to the slightly different results produced by the `cut` function in R and Python, the regression results are also different but very insignificantly.

## Placebo treatment

```python
# subset the data to include only years before 1994
eitc_sub = eitc.where(eitc['year'] <= 1993)

# create a new "after treatment" dummy variable and interaction term
eitc_sub['post91'] = (eitc_sub['year'] >= 1992).astype(int)
```

## `sumstats` Function 

```python
def sumstats(x):
    # the count() function skips NA's by default
    sumtable = x.count().to_frame()
    sumtable['mean'] = round(x.mean(numeric_only = True), 2)
    sumtable['median'] = round(x.median(numeric_only = True), 2)
    sumtable['sd'] = round(x.std(numeric_only = True), 2)
    sumtable['min'] = round(x.min(numeric_only = True), 2)
    sumtable['max'] = round(x.max(numeric_only = True), 2)
    sumtable.columns = ['obs', 'mean', 'median', 'sd', 'min', 'max']
    return sumtable
```
